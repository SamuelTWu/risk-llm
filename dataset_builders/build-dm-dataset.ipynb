# Build Dataset
This is a notebook that asks questions and based on those reposnes builds the dataset. The questions are loaded from a table.

```python
import os
import pandas as pd
import datetime
import numpy as np
import random
```

```python
from transformers import pipeline
#load summarization pipeline (defaults to a pre-trained model like bart-large-cnn)
summarizer = pipeline("summarization")
```

```python
response_dict = {"description":[], "chance-success":[]}
user_input = None
scenarios = pd.read_csv("../data/scenario-dataset.csv")["scenarios"].to_list()
while user_input!="quit":
    attributes = np.random.rand(7)
    text = random.choice(scenarios)
    input_text = text
    num_tokens = len(summarizer.tokenizer.encode(input_text, truncation=False))
    #generate summary
    summary = summarizer(input_text, max_length=int(num_tokens*.6), min_length=int(num_tokens*.3), do_sample=True)
    history = summary[0]['summary_text']
    roll = "SUCCESS" if random.random() < 0.5 else "FAIL"
    output = f"{history}\n(health:{attributes[0]:.2f}\nstrength:{attributes[1]:.2f}\ndexterity:{attributes[2]:.2f}\nperception:{attributes[3]:.2f}\nintelligence:{attributes[4]:.2f}\ncharisma:{attributes[5]:.2f}\nstamina:{attributes[6]:.2f})\n{text}\n{roll}"
    print(output)

    user_input = input()
    
    if user_input == "skip":
        print("skipping prompt")
    else:
        response_dict["description"].append(output)
        response_dict["chance-success"].append(user_input)
    print("\n\n")
    
    
print(response_dict)

df = pd.DataFrame.from_dict(response_dict)
now = datetime.datetime.now()
df.to_csv("../data/dm-dataset-"+now.strftime("%m-%d-%H-%M")+".csv", index=False)
```
