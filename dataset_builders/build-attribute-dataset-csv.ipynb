# Build Dataset
This is a notebook that asks questions and based on those reposnes builds the dataset. The questions are loaded from a table.

```python
import os
import pandas as pd
import datetime
import numpy as np
import random

def is_float(string):
    try:
        float(string)
        return True
    except ValueError:
        return False


response_dict = {"description":[], "dHealth":[], "dStrength":[], "dDexterity":[], "dPerception":[], "dIntelligence":[],"dCharisma":[],"dStamina":[]}
user_input = None
scenarios = pd.read_csv("../data/attribute-scenario-dataset.csv")["scenarios"].to_list()

while user_input!="quit":

    output = random.choice(scenarios) 
    
    print(output+ "\ndHealth|dStrength|dDexterity|dPerception|dIntelligence|dCharisma|dStamina")
    user_input = input()
    user_array = []
    try:
        user_array = list(map(float, user_input.split()))
    except:
        print("An exception occurred")
    
    if all(isinstance(i, float) for i in user_array) and len(user_array)==7:
        response_dict["description"].append(output)
        response_dict["dHealth"].append(float(user_array[0]))
        response_dict["dStrength"].append(float(user_array[1]))
        response_dict["dDexterity"].append(float(user_array[2]))
        response_dict["dPerception"].append(float(user_array[3]))
        response_dict["dIntelligence"].append(float(user_array[4]))
        response_dict["dCharisma"].append(float(user_array[5]))
        response_dict["dStamina"].append(float(user_array[6]))
    else:
        print("Invalid float")
    print("\n\n")
    
    
print(response_dict)

df = pd.DataFrame.from_dict(response_dict)
now = datetime.datetime.now()
df.to_csv("../data/attribute_dataset_"+now.strftime("%m-%d-%H-%M")+".csv", index=False)
```
